```python
categories=pd.read_csv('wiki-topcats-categories.txt',sep="\n",header=None)
```

```python
def num_cat(row):
    cat=row
    return len(cat.split("; ")[1].split(" "))
```

```python
categories=categories[categories.iloc[:,0].apply(lambda x:True if num_cat(x)>3500 else False)]
```

```python
categories_dict=defaultdict(list)
for idx in categories.index:
    cat_and_values=categories[0].loc[idx].split("; ")
    cat_name=cat_and_values[0].split(":")[1]
    categories_dict[cat_name]=list(map(int,cat_and_values[1].split(" ")))
```

```python
from collections import deque
```

```python
def bfs_shortest_path(graph, start):
    explored = set()
    queue = [start]
    shortest_path_dict = {}
    counter_level = 0
    to_visit_list = []
    while queue:
        for node in queue:
            if node not in explored:
                shortest_path_dict[node] = counter_level
                neighbours = graph[node]
                explored.add(node)
                to_visit_list.extend(neighbours)
        queue = []
        queue.extend(to_visit_list)
        to_visit_list = []
        counter_level += 1
    return shortest_path_dict
```

```python
a = bfs_shortest_path(Gtup_directed, categories_dict['English_television_actors'][0])
```

```python
a
```

```python
shortest_path_dictionary_final = {}
for article in tqdm(range(0,len(categories_dict['English_television_actors']))):
    try:
        shortest_path_dictionary_final.update(bfs_shortest_path(Gtup_directed, categories_dict['English_television_actors'][article]))   
    except:
        shortest_path_dictionary_final[article] = -1
```

we've decided to assign value -1 when it's impossible calculating distance between two nodes.
Then we'll replace this value in a reasonable way
